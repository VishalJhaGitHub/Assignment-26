```python
#1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#Ans

#Matplotlib is a Python data visualization library that is used to create static, interactive, and animated visualizations in Python. It is widely used by data scientists, researchers, and engineers to visualize data in various formats.

#Matplotlib is used to create a wide variety of plots, including line plots, scatter plots, bar plots, histograms, and many more. It offers a vast range of customization options that allow users to tailor their plots to their specific needs.

#The Pyplot module is a sublibrary of Matplotlib that provides a convenient interface for creating plots. Some of the most common types of plots that can be plotted using Pyplot include:

#1 - Line plot: A plot that displays data points connected by straight lines. This plot is useful for displaying trends over time.

#2 - Scatter plot: A plot that displays data points as individual dots. This plot is useful for displaying relationships between two variables.

#3 - Bar plot: A plot that displays data as bars. This plot is useful for comparing the values of different categories.

#4 - Histogram: A plot that displays the distribution of data. This plot is useful for displaying the frequency of values in a dataset.

#5 - Pie chart: A plot that displays data as slices of a circle. This plot is useful for displaying the relative proportions of different categories.
```

```python
#2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

#import numpy as np

#np.random.seed(3)

#x = 3 + np.random.normal(0, 2, 50)

#y = 3 + np.random.normal(0, 2, len(x))

#Note: Also add title, xlabel, and ylabel to the plot.

#Ans

#A scatter plot is a type of plot that displays the relationship between two variables by showing the data points as individual dots on a two-dimensional plane. The position of each dot corresponds to the value of the two variables for that data point.

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
#3. Why is the subplot() function used? Draw four line plots using the subplot() function.

#Use the following data:

#import numpy as np

#For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

#For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

#For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

#For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#Ans

#The subplot() function is used in Matplotlib to create multiple plots in a single figure. It takes three arguments: the number of rows of plots, the number of columns of plots, and the index of the current plot.

#Using the subplot() function can be helpful when we want to compare multiple plots side-by-side or when we want to display multiple views of the same data.

import numpy as np
import matplotlib.pyplot as plt
# create data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
# create data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
# create data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
# create data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
# create a figure with two rows of subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
# plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
# plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
# plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
# plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')
# adjust spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)
# display the plot
plt.show()
```

```python
#4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#import numpy as np

#company = np.array(["Apple", "Microsoft", "Google", "AMD"])

#profit = np.array([3000, 8000, 1000, 10000])

#Ans

#A bar plot is a chart that represents data using rectangular bars, where the length or height of the bars indicates the magnitude of the data. Bar plots are useful for comparing different groups or categories of data.

#A horizontal bar plot is a bar plot where the bars are plotted horizontally instead of vertically.

import numpy as np
import matplotlib.pyplot as plt
# create data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
# create a bar plot
plt.bar(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
# create a horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

```python
#5. What is a box plot? Why is it used? Using the following data plot a box plot.

#box1 = np.random.normal(100, 10, 200)

#box2 = np.random.normal(90, 20, 200)

#Ans

#A box plot is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and outliers of the data. The box represents the interquartile range (IQR), which is the range between the 25th and 75th percentiles of the data. The line inside the box represents the median, and the whiskers represent the range of the data, excluding outliers. Outliers are data points that are more than 1.5 times the IQR away from the box.

import numpy as np
import matplotlib.pyplot as plt

# Create data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
# Combine the data into a list
data = [box1, box2]
# Create the boxplot
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_title('Box Plot')
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
plt.show()
```
